## Introduzindo scipy

```python
from scipy.optimize import curve_fit 
```

```python
def my_power_law(x,a,b): 
    return a*x**b
```

```python
par, cov = curve_fit(my_power_law,spotify['s'],spotify['m'], maxfev=2000)
```

```python
par
```

```python
plt.plot(spotify['s'],spotify['m'], '.k')
yfit = my_power_law(spotify['s'],par[0], par[1])

plt.plot(spotify['s'], yfit, color = "red")
```
